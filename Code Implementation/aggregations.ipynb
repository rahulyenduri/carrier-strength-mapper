```python
gdf_dropped = gdf.drop(columns=['APN', 'FLAG', 'MAIL_STREE',
       'MAIL_CITY', 'SITUS_STRE', 'SITUS_CITY', 'STREET_NUM', 'STREET_NAM',
       'STREET_TYP', 'CITY', 'ZIP_CODE', 'CLASS_CODE', 'MULTIPLE',
       'SUBDIVISIO', 'ACREAGE', 'RECORDER_M', 'BOOK', 'PAGE', 'MAP_BOOK_P',
       'LOT_TYPE', 'LOT', 'CAME_FROM', 'TAX_RATE_A', 'LAND', 'STRUCTURES',
       'PRIMARY_OW', 'ALL_OWNER_', 'SHAPESTAre', 'SHAPESTLen', 'geometry'])
```

```python
gdf_dropped
```

```python
grouped_gdf = gdf_dropped.groupby(['OBJECTID','DBA']).agg({
    'Technology': 'max'
}).reset_index()
```

```python
grouped_gdf
```

```python
grouped_gdf['max_column3_over_partition'] = grouped_gdf.groupby('OBJECTID')['Technology'].cummax()
```

```python
grouped_gdf
```

```python
grouped_gdf['Concatenated'] = grouped_gdf['DBA'] + ' : ' + (grouped_gdf['Technology']).astype(str)
```

```python
grouped_gdf
```

```python
grouped_gdf['ConcatenatedValue'] = grouped_gdf.groupby(grouped_gdf.columns[0:1].tolist())['Concatenated'].transform(' , '.join)
grouped_gdf['Available_Technology'] = grouped_gdf.groupby(grouped_gdf.columns[0:1].tolist())['DBA'].transform(' , '.join)
```

```python
grouped_gdf
```

```python
grouped_gdf.Available_Technology.unique()
```

```python
result_df = grouped_gdf.loc[grouped_gdf.groupby('OBJECTID')['Technology'].idxmax()]
```

```python
result_df
```

```python
len(result_df)
```

```python
filt = result_df[result_df.OBJECTID == 193566]
```

```python
result_df = result_df.drop(columns=['Concatenated','Technology'])
```
